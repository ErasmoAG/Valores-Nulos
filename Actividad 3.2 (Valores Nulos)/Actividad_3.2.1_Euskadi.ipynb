```python
#Importamos librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .csv sin indice
data1= pd.read_excel('Euskadi.xlsx')
data1
```

```python
#Identificar valores nulos por columna
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Identificar valores nulos para dataframe
valores_nulos=data1.isnull().sum().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
data2=data1.copy()
```

```python
#Sustitucion de valores nulos por un valor de texto dado que no proporcionan una descripcion, llenarlo con una descripcion diciendo que no esta disponible sirve para no dejar los valores nulos.
data2["description"]=data2["description"].fillna("No description available")
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustitucion de valores nulos por un valor de texto, al igual que la columna anterior, la use para dar entender que la informacion no fue proporciona y va acorde con el tipo de valor que tiene la columna.
data2["neighborhood_overview"]=data2["neighborhood_overview"].fillna("No overview available")
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustitucion de valores nulos por un valor de texto, al igual que la columna anterior opte por sustituir el valor nulo por desconocido.
data2[["host_response_time"]]=data2[["host_response_time"]].fillna("Unknown")
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos con promedio o media, se eligio este metodo para estas columnas porque son numéricas, y la media es adecuada para representar valores faltantes sin distorsionar el análisis.
data2[["host_acceptance_rate", "review_scores_rating", "review_scores_cleanliness", "review_scores_checkin", "review_scores_communication", "review_scores_location", "review_scores_value"]]=data2[["host_acceptance_rate", "review_scores_rating", "review_scores_cleanliness", "review_scores_checkin", "review_scores_communication", "review_scores_location", "review_scores_value"]].fillna(round(data1[["host_acceptance_rate", "review_scores_rating", "review_scores_cleanliness", "review_scores_checkin", "review_scores_communication", "review_scores_location", "review_scores_value"]].mean(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustitucion de valores nulos por un valor de texto, al igual que la columna anterior, y esta vez por dos columnas opte por sustituir el valor nulo por desconocido, y dado que es acorde y se puede aplicar a diferentes columnas, asi lo hice.
data2[["host_neighbourhood"]]=data2[["host_neighbourhood"]].fillna("Unknown")
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustitucion de valores nulos con valores no nulos hacia atras, decidi ocupar este metodo dado por el hecho de que hay varios anfritiones con varias propiedades y algunas si estan marcadas como verrdaderos y otros como falso y al ocupar este metodo se puede llenar con el hecho que lo puede ser o no, y dado con los valores de la tabla es una buena susticion para los valores nulos. 
data2[["host_is_superhost"]]=data2[["host_is_superhost"]].fillna(method="bfill")
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustitucion de valores nulos por valores no nulos hacia adelante, ocupe este metodo y a varias columnas ya que visualizando los valores, me di cuenta el valor adelante de el suele ser el mismo o similar al de la columna y queda perfecto como sustitucion de los valores nulos 
data2[["bathrooms_text"]]=data2[["bathrooms_text"]].fillna(method="ffill")
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustitucion de valores nulos con promedio o  media, decidi ocupar este metodo y dada la similitud de columnas, dado que los valores de cada columna suelen estar en casi el mismo rango de valor, asi que al sacar el promedio de cada columna y sustituirlo por lo valores nulos es una buena aproximacion de los valores.
data2[["bedrooms", "beds", "price"]]=data2[["bedrooms", "beds", "price"]].fillna(round(data1[["bedrooms", "beds", "price"]].mean(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
##Sustitucion de valores nulos por un valor de texto en concreto, al igual que los anteriores este lo escogi dado que si no tiene un valor parece razonable pensar que no esta disponible.
data2["has_availability"] = data2["has_availability"].fillna("unavailable")
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustituicion por mediana ya que las columnas pueden tener variabilidad y este metodo es más resistente a valores extraños o discrepancias que puede tener al ser columnas de reseñas,
data2[["reviews_per_month"]]=data2[["reviews_per_month"]].fillna(round(data1[["reviews_per_month"]].median(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Identificar valores nulos para dataframe
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Identificar valores nulos para dataframe
valores_nulos=data2.isnull().sum().sum()
valores_nulos
```
